# Linear Regression with Single Variable

## Import required libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Read data from .csv file

```python
data = pd.read_csv("../data/income_by_year.csv")
```

```python
data.head()
```

```python
data.describe()
```

```python
data.info()
```

## Data Visualization
* The data suitable for linear model

```python
plt.plot(data.year, data["per capita income (US$)"])
plt.title("Per capita income in US")
plt.xlabel("Year")
plt.ylabel("Income in (US$)")
plt.show()
```

## Import sklearn for create linear model

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

## Train model with the data

```python
model.fit(data[['year']], data['per capita income (US$)'])
```

## Visualize Prediction and Real values

```python
plt.plot(data.year, data.iloc[:, 1], c="blue")
plt.plot(data.year, model.predict(data[['year']]), c="red")

plt.title("Per Capita Income (US$)")
plt.xlabel("Year")
plt.ylabel("Income (US$)")
plt.legend(["Real", "Prediction"])

plt.show()
```

## Let's predict the future

```python
model.predict([[2025]])
```
